## **1. Getting and Knowing Your Data**

```
#Import the necessary libraries
 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
from google.colab import files


movies = pd.read_csv('Movies Box Office Data Set.csv')      
print(movies)  
```

```

movies.head(40)
```

```

movies[movies.columns[3]] = movies[movies.columns[3]].replace('[\$,]', '', regex=True).astype(float)
movies[movies.columns[4]] = movies[movies.columns[4]].replace('[\$,]', '', regex=True).astype(float)
movies[movies.columns[6]] = movies[movies.columns[6]].replace('[\$,]', '', regex=True).astype(float)
movies
```

## **2. Stats**

```
grouped_single = movies.groupby('Year').agg({'Worldwide': ['sum', 'std', 'var'],'Domestic': ['sum', 'std', 'var'],'Foreign': ['sum', 'std', 'var']})
grouped_single
```

## **3. Correlation**

```
fig, ax = plt.subplots(figsize=(10,10))
df=pd.DataFrame(movies,columns=['Worldwide','Domestic','Foreign','Year'])
plt.scatter(df.Year,df.Worldwide)
plt.xlabel("Year")
plt.ylabel("Gross in Billion Dollars")
```

## **4. Visualization**

```
#This exercise was created based on the tutorial and documentation from Seaborn
#The dataset being used is tips from Seaborn.

#Import the necessary libraries
import pandas as pd

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("white")
```

```
 
histo = sns.histplot(y ="Worldwide", x ="Year", data = movies);
histo.set(ylabel = 'Gross in Billion Dollars', xlabel = 'Year', title = "Worldwide")
```

```
histo = sns.histplot(y ="Domestic", x ="Year", data = movies);
histo.set(ylabel = 'Gross in Billion dollars', xlabel = 'Year', title = "Domestic")
```

```

histo = sns.histplot(y ="Foreign", x ="Year", data = movies);
histo.set(ylabel = 'Gross in Billion dollars', xlabel = 'Year', title = "Foreign")
```
